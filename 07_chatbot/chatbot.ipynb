```python
import random

print("Chatbot : Hello! I am your simple chatbot. Type 'bye' to exit.\n")

# Responses dictionary
responses = {
    "hello": ["Hi!", "Hello ", "Hey there!"],
    "how are you": ["I'm doing great ", "All good here!", "I'm fine, thanks!"],
    "your name": ["I'm a Python chatbot ", "You can call me Bot!"],
    "help": ["I can chat with you Try saying hello, how are you, or ask my name."],
    "bye": ["Goodbye ", "See you later!", "Take care "]
}

def get_response(user_input):
    user_input = user_input.lower()

    for key in responses:
        if key in user_input:
            return random.choice(responses[key])

    return "Sorry, I don't understand that 🤔"

# Chat loop
while True:
    user_input = input("You: ")

    reply = get_response(user_input)
    print("Bot:", reply)

    if "bye" in user_input.lower():
        break
```

below is an example of a knowledge baesd chatbot which answers simple questions relevant to the kb stored,
we can also contain factual information in the kb too

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Knowledge Base
questions = [

    "what is python",
    "what is ai",
    "what is machine learning",
    "what is nlp",
    "who developed python",
    "what is data science",
    "what is deep learning"

]

answers = [

    "Python is a programming language.",

    "AI stands for Artificial Intelligence.",

    "Machine Learning is a subset of AI.",

    "NLP stands for Natural Language Processing.",

    "Python was developed by Guido van Rossum.",

    "Data Science is the process of extracting insights from data.",

    "Deep Learning is a subset of Machine Learning."

]

print("🤖 Chatbot: Type 'bye' to exit.\n")

while True:

    # User input
    user_input = input("You: ").lower()

    # Exit condition
    if user_input == "bye":

        print("🤖 Chatbot: Goodbye!")
        break

    # Combine questions with user input
    all_sentences = questions + [user_input]

    # TF-IDF Vectorization
    vectorizer = TfidfVectorizer()

    tfidf_matrix = vectorizer.fit_transform(all_sentences)

    # Cosine Similarity
    similarity_scores = cosine_similarity(

        tfidf_matrix[-1],
        tfidf_matrix[:-1]

    )

    # Find most similar question
    index = similarity_scores.argmax()

    # Get similarity score
    score = similarity_scores[0][index]

    # Threshold checking
    if score > 0.3:

        response = answers[index]

        print("🤖 Chatbot:", response)

    else:

        print("🤖 Chatbot: Sorry, I don't understand.")
```
