# What is Tokenization?
Tokenization means breaking text into smaller pieces called tokens.

Example:

"NLP is fun"

becomes:

["NLP", "is", "fun"]

 Tokenization - Word and Sentence Tokenisation

 Word Tokenisation : Sentences are broken down into simple individual words
 Sentence Tokenisation : Paragraphs are broken into individual sentences


```python
from nltk.tokenize import word_tokenize
text="NLP is fun!"
tokens=word_tokenize(text)
print("The tokens are:" , tokens)
# Note : punctuation is also treated as token
```

```python
from nltk.tokenize import sent_tokenize
para="NLP is fun. It is easy to learn nlp with me."
sentences=sent_tokenize(para)
print("The tokens are:" , sentences)
#Note : space must be provided between sentences then only sentence tokennization works
```

# What are Stopwords?
Stopwords are commonly used words like:

is
the
am
are

These usually add little meaning in NLP tasks.

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
text="NLP is easy to learn with me"
tokens=nltk.word_tokenize(text)
filtered=[]
for word in tokens:
    if word.lower() not in stop_words:
        filtered.append(word)
print("the filtered tokens are:",filtered)
```

# Frequency Distribution
Why Important?
Frequency analysis helps:

spam detection
keyword extraction
trend analysis
search engines

```python
from nltk.probability import FreqDist
text="NLP is fun and NLP is powerful"
tokens=nltk.word_tokenize(text)
flist=FreqDist(tokens)
print("the frequency list of tokens are:",flist)
```

```python
print(flist.most_common(3))
# finds the most common 3 words
```
