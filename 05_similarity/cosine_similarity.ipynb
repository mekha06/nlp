```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Step 1: Read sentences from file
file_path = "sentences.txt"

with open(file_path, "r") as file:
    sentences = file.readlines()

# Clean sentences
sentences = [s.strip() for s in sentences if s.strip()]

# Step 2: User input
user_input = input("Enter your sentence: ")

# Add user input to list (so vectorizer can compare)
all_sentences = sentences + [user_input]

# Step 3: TF-IDF Vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(all_sentences)

# Step 4: Cosine similarity (user sentence vs all others)
similarity_scores = cosine_similarity(
    tfidf_matrix[-1], tfidf_matrix[:-1]
)

# Step 5: Find most similar sentence
most_similar_index = similarity_scores.argmax()
most_similar_sentence = sentences[most_similar_index]

# Output results
print("\nMost similar sentence is:")
print(most_similar_sentence)

print("\nSimilarity score:", similarity_scores[0][most_similar_index])
```
