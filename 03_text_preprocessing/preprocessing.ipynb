# Text Preprocessing

Text preprocessing is the process of cleaning and transforming raw text into a format suitable for NLP tasks.

Preprocessing helps improve:
- accuracy
- efficiency
- consistency

Applications:
- Chatbots
- Sentiment Analysis
- Search Engines
- Spam Detection

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import string
```

```python
text = "Natural Language Processing is AMAZING! NLP helps computers understand languages easily."
```

```python
#already discussed about tokenisation....
tokens = word_tokenize(text)

print(tokens)
```

Lowercasing
Why Important?
Computers treat:

NLP
nlp
Nlp
as DIFFERENT words.
Lowercasing normalizes them.

```python
lower_words=[word.lower() for word in tokens]
print(lower_words)
```

Removing Punctuation
Why Remove Punctuation?

Usually punctuation:

adds noise
increases vocabulary unnecessarily

```python
clean_words=[]
for word in lower_words:
    if word not in string.punctuation:
        clean_words.append(word)
print(clean_words)
```

```python
# stopword removal(discussed already)
stop_words = set(stopwords.words('english'))

filtered_words = []

for word in clean_words:
    if word not in stop_words:
        filtered_words.append(word)

print(filtered_words)
```

What is Stemming?

Stemming reduces words to ROOT form.

Examples:
Original	Stem
playing	play
studies	studi
running	run

Stemming may produce:

studies → studi

which is NOT a real word.

Because stemming uses rule-based chopping.
Thus less accurate with generation of invalid vocab

```python
stemmer = PorterStemmer()

stemmed_words = []

for word in filtered_words:
    stemmed_words.append(stemmer.stem(word))

print(stemmed_words)
```

Lemmatization
What is Lemmatization?

Lemmatization converts words into meaningful dictionary forms.

Examples:

Original	Lemma
studies	study
running	run
better	good

lemmatization produces accurate results than stemming
you can evidently see the difference in the ouputs of stemming and lemmatization

| Feature  | Stemming             | Lemmatization |
| -------- | -------------------- | ------------- |
| Speed    | Faster               | Slower        |
| Accuracy | Lower                | Higher        |
| Output   | May not be real word | Real word     |


```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = []

for word in filtered_words:
    lemmatized_words.append(lemmatizer.lemmatize(word))

print(lemmatized_words)
```
