````python
import nltk
from nltk.translate import IBMModel1
from nltk.translate import AlignedSent

# Parallel corpus (training data)
bitext = [

    AlignedSent(
        ['je', 'suis', 'etudiant'],
        ['i', 'am', 'student']
    ),

    AlignedSent(
        ['j', 'aime', 'python'],
        ['i', 'love', 'python']
    ),

    AlignedSent(
        ['bonjour'],
        ['hello']
    ),

    AlignedSent(
        ['merci'],
        ['thankyou']
    )
]

# Train IBM Model 1
ibm1 = IBMModel1(bitext, 10)

# User input
sentence = input(
    "Enter French-like sentence: "
).lower()

# Tokenization
words = sentence.split()

translated_sentence = []

# Translation process
for word in words:

    best_translation = None
    best_probability = 0

    # Check if word exists in translation table
    if word in ibm1.translation_table:

        # Get all possible translations
        for eng_word, prob in ibm1.translation_table[word].items():

            # Find highest probability translation
            if prob > best_probability:

                best_probability = prob
                best_translation = eng_word

    # Add translated word
    if best_translation:

        translated_sentence.append(best_translation)

    else:

        translated_sentence.append(word)

# Join translated words
result = " ".join(translated_sentence)

# Output
print("\nTranslated Sentence:")
print(result)
````
