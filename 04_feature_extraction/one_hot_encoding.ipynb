# Why Text Representation is Needed

Machine Learning models cannot understand raw text directly.

Example:

"I love NLP"

Computers must convert words into numerical format before processing.

One Hot Encoding is one of the simplest text representation techniques in NLP.

Sentence:

"I love NLP"

Vocabulary:

["I", "love", "NLP"]

One Hot Representation:

Word	Vector
I	    [1,0,0]
love	[0,1,0]
NLP	    [0,0,1]

Each word gets:

unique vector
only one position = 1
rest = 0

That’s why it is called:

One Hot Encoding

One Hot Encoding cannot understand meaning.

Example:

king
queen
man
woman

All are treated as unrelated vectors.

This is why advanced embeddings exist later:

Word2Vec
GloVe
BERT embeddings

# Advantages of One Hot Encoding

- Simple to understand
- Easy to implement
- Useful for small vocabularies
- Beginner-friendly

# Disadvantages

- Large vectors for large vocabulary
- Sparse representation
- No semantic meaning
- Cannot capture word relationships

```python
#manual way of doing one-hot vectors

sentence="I love nlp"
sent=sentence.split()
vocab=list(set(sent))
print("the vocabulary : ", vocab)

one_hot={}

for i,word in enumerate(vocab):
    vector=[0]*len(vocab)
    vector[i]=1
    one_hot[word]=vector

print("one_hot_vectors are\n" , one_hot)
```

```python
#better version preserving the word order ,in the above case since we use set to remove duplicates it doesnt preserve the word order ,so we use a list appending method 
sentence = "I love nlp"

words = sentence.split()

vocabulary = []

for word in words:
    if word not in vocabulary:
        vocabulary.append(word)

print("Vocabulary:", vocabulary)

one_hot_vectors = {}

for i, word in enumerate(vocabulary):

    vector = [0] * len(vocabulary)

    vector[i] = 1

    one_hot_vectors[word] = vector

print("\nOne Hot Vectors:\n")

for word, vector in one_hot_vectors.items():
    print(word, ":", vector)
```
