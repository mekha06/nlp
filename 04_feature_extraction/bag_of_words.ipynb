# Bag of Words (BoW)

Bag of Words is a text representation technique used in NLP and Machine Learning.

It converts text into numerical vectors based on word frequency.

Important:
- Grammar is ignored
- Word order is ignored
- Only frequency matters

Sentence:

"I love NLP"

Another sentence:

"I love Machine Learning"

Vocabulary becomes:

['I', 'love', 'NLP', 'Machine', 'Learning']

Now each sentence becomes a frequency vector.

Example Representation
Word	Sentence 1	Sentence 2
I	        1	         1
love	    1	         1
NLP	        1	         0
Machine	    0	         1
Learning	0	         1

IMPORTANT DIFFERENCE
One Hot Encoding

Represents:

individual words

Bag of Words

Represents:

entire documents/sentences

VERY important distinction.

```python
sentence = input("Enter the text: ")

words = sentence.split()

vocab = []

for word in words:

    if word not in vocab:

        vocab.append(word)

print("The vocabulary is:", vocab)

bow_vector = []

for word in vocab:

    bow_vector.append(words.count(word))

print("\nBoW Vector:")

print(bow_vector)
```

```python
sentences = []

n = int(input("Enter number of sentences: "))

for i in range(n):

    sentence = input(f"Enter sentence {i+1}: ").lower()

    sentences.append(sentence)

# Create Vocabulary

vocab = []

for sentence in sentences:

    words = sentence.split()

    for word in words:

        if word not in vocab:

            vocab.append(word)

print("\nVocabulary:\n")

print(vocab)

# Create BoW Matrix

bow_vectors = []

for sentence in sentences:

    vector = []

    words = sentence.split()

    for word in vocab:

        vector.append(words.count(word))

    bow_vectors.append(vector)

# Display BoW Vectors

print("\nBoW Matrix:\n")

for i, vector in enumerate(bow_vectors):

    print(f"Sentence {i+1}:")

    print(vector)
```
